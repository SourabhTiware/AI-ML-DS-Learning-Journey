### Lec 04: Series => 1D labelled array
- First type of DataStructure in pandas is "Series"

```python
# Define Series using py_list

s = pd.Series([1,2,3,4,5])
print(s)
print(type(s))
```

```python
# access element of series 
# first method using Index 

print(s[0])
print(s[4])
```

```python
# but also we can assign index => lable value for series like

s = pd.Series([1,2,3,4,5],index = ["Sourabh","Mina","Raj","Jaya","Dhanashree"])
print(s)
```

```python
# now we can access series element using the labelled. 

print(s["Sourabh"])
print(s["Jaya"])

# also we can access element using index 
print(s[0]) # also for now just ignore the warning. 
```

### Lec 05: Series Properties

- Homogenous
- Vectorized Operations
- Handle Missing values with NaN
- Mutable values, immutable size

```python
# Homogenous 
s = pd.Series([1,2,3,4,5,"xzy"],index = ["Sourabh","Mina","Raj","Jaya","Dhanashree",32])
print(s)
# Observe dtype of current series and above cell seires. here I can added "xzy" string into the seires
# then pandas change it's type into to object. Means it's Homogenous means store all types of data. 

# print(s["Sourabh"])
```

```python
# Vectorized Operation 
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([10,20,30,40,50])

print(s1 + s2)
```

```python
# Handle Missing values with NaN => Learn in upcoming Lecture

# Mutable values, immutable size

s1 = pd.Series([1,2,3,4,5])
 # Mutuable values 
s1[0] = 47
print(s1)

# but immutable in size means we want to drop or add a element in series we want to create another new series. 

s1.drop(0)
print(s1) # It's not working but following lines are working becuse of assign new changes in another variable.  

changed_s1 = s1.drop(0)
print(changed_s1)
```
