```python
import json
```

```python
def load_data(file_name):
    with open(file_name, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

```python
# Clean & sturcuture the data

def clean_data(data):
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}
    clean_data = []
    unique_set = set()
    for user in data:
        raw_rating = str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        #Handle missing value
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        # Deduplication 
        if user["name"].strip() in unique_set:
            continue
        unique_set.add(user["name"])
        clean_data.append(user)
    return clean_data  
```

```python
final_data = clean_data(data)
```

```python
print(final_data)
```

```python
# Calculate the avarage rating of user for the particular product. 
def cal_average_rating(final_data):
    sum = 0
    count = 0
    for user in final_data:
        sum += float(user["rating"])
        count += 1
    
    average_rating = sum / count
    return average_rating
```

```python
total_average = cal_average_rating(final_data)
print(f"total average rating of product = {total_average}")
```

```python
# give insight getting poor rating. 

def cal_poor_rating(final_data):
    sum = 0
    count = 0
    length_user = len(final_data)

    for user in final_data:
        if(float(user["rating"]) < 3):
            count += 1

    return count / length_user * 100
```

```python
poor_rating = cal_poor_rating(final_data)
print(f"Poor rating of people is {poor_rating}%")
```

```python
# Recommendation system. 

def recommendation_sys(final_data):
    recommendation = []

    for user in final_data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]
        if(float(user["rating"]) >= 4):
            curr_recomm["brand"] = "apple"
        else:
            curr_recomm["brand"] = "samsung"

        recommendation.append(curr_recomm)
    return recommendation    
```

```python
recommendation_for_user = recommendation_sys(final_data)
print(recommendation_for_user)
```
